## Merging Datasets

```python
order_merge = pd.merge(orders, order_details, on="order_id")

order_merge
```

```python
pizza_merge = pd.merge(pizzas, pizza_types, on="pizza_type_id")

pizza_merge.head(10)
```

```python
pizza_data = pd.merge(order_merge, pizza_merge, on="pizza_id")

pizza_data.head(10)
```

## Total Revenue

```python
total_revenue = (pizza_data["price"] * pizza_data["quantity"]).sum()

print(f"The total revenue is {total_revenue}")
```

## Peak Hours of Sale

```python
pizza_data.info()
```

```python
pizza_data
```

```python
#combining the date and time column into a datetime column for easy calculation

pizza_data["order_timestamp"] = pd.to_datetime(pizza_data["date"].astype(str)+" "+pizza_data["time"].astype(str))

pizza_data.head()
```

```python
#converting the order timestamp column to datetime

pizza_data["date"]= pd.to_datetime(pizza_data["date"])

pizza_data.info()
```

```python
pizza_data = pizza_data.drop(columns=["date", "time"])

pizza_data.head()
```

```python
peak_hours = pizza_data["order_timestamp"].dt.hour.value_counts()

peak_hours
```

```python
print(f"The peak hours are {peak_hours.index[:5]}")
```

The peak hours are 12PM, 13PM, 18PM, 17PM, 19PM. This shows that the highest sales are during the midday, which coincide with office breaktime and students lunchtime. The next peak is during closing hours, when workers are closing from work, and students are returning from school.

## Total Sales Made on Each Day of the Week and Day with the Highest Sale

```python
#creating a day od the week column

pizza_data["day_of_week"] = pizza_data["order_timestamp"].dt.day_name()

pizza_data.head()
```

```python
#creating a revenue column to show the revenue for each row
pizza_data["revenue"] = pizza_data["quantity"] * pizza_data["price"]

pizza_data.head()
```

```python
daily_sales = pizza_data.groupby("day_of_week")["revenue"].sum()

daily_sales
```

```python
#Rearranging the index serially
use_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

daily_sales = daily_sales.reindex(use_order)
daily_sales
```

```python
print(f"The total sales made on each day of the of the week is \n{daily_sales}")
```

```python
#Displaying the day with the highest sale

print(f"The total sales made on each day of the of the week is \n{daily_sales}")
```

#### Plotting the Data

```python
#Converting to dataframe for plotting
plot_data = daily_sales.reset_index()

plot_data
```

```python
sns.barplot(x=plot_data["day_of_week"], y=plot_data["revenue"], data=plot_data, palette="viridis", hue=plot_data["day_of_week"])

plt.title("Revenue per Day", fontsize=15)
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.xticks(rotation = 90)

plt.savefig('daily_revenue.png', dpi=300, bbox_inches='tight')

plt.show()
```

## The Sales Made in Each Month

```python
pizza_data.head()
```

```python
pizza_data["month"] =  pizza_data["order_timestamp"].dt.month_name()

pizza_data
```

```python
monthly_sales = pizza_data.groupby("month")["revenue"].sum()

monthly_sales
```

```python
#Rearranging the month serially
month_order = ['January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December']

monthly_sales = monthly_sales.reindex(month_order)

monthly_sales
```

```python
print(f"The month with the highest revenue is {monthly_sales.idxmax()}.\nThe month with the lowest revenue is {monthly_sales.idxmin()}")
```

```python
#converting to dataframes for plotting
monthly_plot = monthly_sales.reset_index()

monthly_plot
```

#### Plotting the Information

```python
sns.barplot(x=monthly_plot["month"], y=monthly_plot["revenue"], data=monthly_plot, palette="rocket", hue=monthly_plot["month"], legend=False)

plt.title("Monthly Sales", fontsize=15)
plt.xlabel("Months", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.xticks(rotation = 90)

plt.show()
```

The data shows a peak in June followed by a significant decline in October. This suggests a seasonal demand pattern where summer social gatherings drive higher volume, while the transition into autumn routines and holiday saving leads to a cooling of sales in October.
